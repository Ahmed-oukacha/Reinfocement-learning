```python
import numpy as np
import random
```

```python
grille = np.array([
    [0, 0, 0, 0, 0],
    [0, -10, 0, -10, 0],
    [0, 0, 0, 0, 0],
    [0, 0, 10, 0, 0],
    [0, 0, 0, 0, 0]
])
```

```python
actions = [(0, 1), (0, -1), (1, 0), (-1, 0)]
```

```python
def get_next_state(state, action):
    new_state = (state[0] + actions[action][0], state[1] + actions[action][1])
    if 0 <= new_state[0] < grille.shape[0] and 0 <= new_state[1] < grille.shape[1]:
        return new_state
    return state
```

```python
def jouer_une_partie():
    state = (0, 0)  
    done = False
    total_reward = 0
    while not done:
        action = random.randint(0, 3) 
        next_state = get_next_state(state, action)
        reward = grille[next_state[0], next_state[1]] - 1  # CoÃ»t de dÃ©placement de -
        total_reward += reward

        print(f"DÃ©placement vers {next_state}, RÃ©compense: {reward}, Score total: {total_reward}")
        
        if grille[next_state[0], next_state[1]] == 10:
            print("ðŸŽ‰ TrÃ©sor trouvÃ© ! Fin de la partie.")
            done = True
            return done
        elif grille[next_state[0], next_state[1]] == -10:
            print("ðŸ’€ PiÃ¨ge ! Partie terminÃ©e.")
            done = False
            
        else:
            state = next_state
        return False    
```

```python
for i in range(10):
    print(f"partie {i}")
    a = jouer_une_partie()
    if a :
      break
    print("----------------")
```
