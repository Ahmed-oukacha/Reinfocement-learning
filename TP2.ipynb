```python
import gymnasium as gym

env = gym.make("FrozenLake-v1", render_mode="human")
env.reset()
```

```python
print(f"Espace d'action : {env.action_space}")
print(f"Espace d'observations : {env.observation_space}")


for _ in range(2):
    action = env.action_space.sample()
    observation, reward, done, _, _ = env.step(action)
    print(f"Action : {action}, Observation : {observation}, reward : {reward}")
    if done: 
        env.reset()
env.close()
    
```

```python
import numpy as np
q_table = np.zeros((env.observation_space.n, env.action_space.n))
```

```python
print("l'initialisation q_table")
print(q_table)
```

```python
alpha = 0.1
gamma = 0.99
epsilon = 1
epsilon_decay = 0.995
num_episodes = 5000
eps = 0.01
num_states = env.observation_space.n  
num_actions = env.action_space.n  
Q_table = np.zeros((num_states, num_actions))


for episode in range(num_episodes):
    state = env.reset()[0]  
    done = False

    while not done:
        
        if np.random.uniform(0, 1) < epsilon:
            action = env.action_space.sample()  
        else:
            action = np.argmax(Q_table[state])  

        
        next_state, reward, done, _, _ = env.step(action)

        
        Q_table[state, action] = Q_table[state, action] + alpha * (
            reward + gamma * np.max(Q_table[next_state]) - Q_table[state, action]
        )

        
        state = next_state
    epsilon = max(eps, epsilon * epsilon_decay)

    
    


print("Q-Table après entraînement :", Q_table)
```

```python
num_test_episodes = 100
success_count = 0  

for episode in range(num_test_episodes):
    state = env.reset()[0] 
    done = False

    while not done:
        action = np.argmax(Q_table[state])  
        next_state, reward, done, _, _ = env.step(action)

        if done and reward == 1: 
            success_count += 1

        state = next_state  

success_rate = (success_count / num_test_episodes) * 100
print(f"Taux de réussite de l'agent : {success_rate:.2f}% sur {num_test_episodes} épisodes.")
```
